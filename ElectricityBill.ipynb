```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score

```

```python
df=pd.read_csv("Electricity Bill.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
X=df[['NumOccupants', 'Temperature', 'Area', 'AC_hours', 'AC_no',
       'AC_star_rating', 'efficiency_factor', 'Washing_hours', 'Geyser_hours',
       'Fridge_hours']]
Y=df[['Price']]
```

```python
X
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y, test_size=0.2, random_state=7)
```

```python
X_test
```

```python
Y_test
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
prediction=model.predict(X_train)
print("MAE:", mean_absolute_error(prediction, Y_train))
print("R2 Score:", r2_score(prediction,Y_train))

```

```python
prediction=model.predict(X_test)
print("MAE:", mean_absolute_error(prediction, Y_test))
print("R2 Score:", r2_score(prediction,Y_test))
```

```python
import joblib
joblib.dump(model, 'bill_predictor_model.pkl')
```
